## Plotting example Lyman-α profiles

In this notebook, we plot example Lyman alpha profiles for sources with all baseline types (constant, linear, damped absorption) and line multiplicities (single or double peaked).

```python
import matplotlib.pyplot as plt
from pathlib import Path

# List of idens to be plotted
idens = [
    'MACS0416S-P5061',
    'RXJ1347-SP3508',
    'MACS0940-SP317',
    'A370-SP8253',
    'RXJ1347-P4556',
    'RXJ1347-SP3179'
]

# First pass: find the first valid image to get aspect ratio
img_aspect_ratio = None
for iden in idens:
    cluster, source = iden.split('-', 1)
    plot_dir = Path(f"../muse_data/{cluster}/plots/{source}/LYALPHA_fit_{SPEC_TYPE}.png")
    if plot_dir.exists():
        img = plt.imread(plot_dir)
        img_aspect_ratio = img.shape[1] / img.shape[0]  # width / height
        break

if img_aspect_ratio is None:
    print("Warning: Could not find any valid images to determine aspect ratio")
    img_aspect_ratio = 1.0

# Create figure with appropriate dimensions based on image aspect ratio
# 2 rows, 3 columns. Each subplot should be 5 inches wide (15/3).
# If image aspect ratio is ar, each plot should be 5/ar inches tall.
# With 2 rows, total height = 2 * (5/ar) = 10/ar
fig, ax = plt.subplots(2, 3, figsize=(15, 10 / img_aspect_ratio), facecolor='white')
axs = ax.flatten()

# Set aspect ratio for each axis
for axis in axs:
    axis.set_aspect(img_aspect_ratio)

nax = 0
for iden in idens:
    # Parse cluster and source from iden
    cluster, source = iden.split('-', 1)
    
    # Bring up the saved plot of this source for visual inspection
    plot_dir = Path(f"../muse_data/{cluster}/plots/{source}/LYALPHA_fit_{SPEC_TYPE}.png")
    if plot_dir.exists():
        img = plt.imread(plot_dir)
        axs[nax].imshow(img)
        axs[nax].axis('off')
        nax += 1
    else:
        print(f"  Plot not found at {plot_dir}")

plt.subplots_adjust(left=0.01, right=0.99, top=0.99, bottom=0.01, wspace=-0.25, hspace=0.05)
plt.savefig(f"/home/james/Documents/my_papers/lensed_laes/paper2/figures/example_lya_profiles.pdf", dpi=500,
            bbox_inches='tight')
plt.show()
```
